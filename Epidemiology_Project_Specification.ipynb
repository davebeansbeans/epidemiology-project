```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def SIR(y, t, beta):
    # returns RHS of the SIR model
    S, I, R = y

    mu = 0.1
    
    Sdot = -beta * S * I
    Idot = beta * S * I - mu * I
    Rdot = mu * I
    
    return Sdot, Idot, Rdot

# set up time grid for the solution
tf = 175
Nsteps = 1000
t = np.linspace(0, tf, Nsteps+1)

# set initial conditions and put into array y0
N = 10**4
I0 = 3
S0 = N - I0
R0 = 0
y0 = np.array([S0, I0, R0])

# call ODE solver
y_sol = odeint(SIR, y0, t, (2*(10**-5),))

# extract S, I, R from the solution array y
S = y_sol[:,0]
I = y_sol[:,1]
R = y_sol[:,2]
Imax = I.max()

# setting up figures and subplots
fig1, (ax11, ax12) = plt.subplots(nrows=2,figsize=(6,10))
fig1.subplots_adjust(hspace=0.3)

# plot the time evolution of S, I, R
ax11.plot(t, S, label = "S")
ax11.plot(t, I, lw = 3, label = "I")
ax11.plot(t, R, label = "R")
ax11.plot(t[np.where(I == Imax)], Imax, 'ko', markersize = 8)

# labels etc
ax11.set_xlabel("time/days", fontsize = 14)
ax11.set_ylabel("S, I, R", fontsize = 14)
ax11.set_title("Deterministic SIR model (Baseline case)", fontsize = 16)
ax11.legend(fontsize = 14)

# plot the evolution in the S,I phase plane
ax12.plot(S, I)
ax12.plot(S[np.where(I == Imax)], Imax, 'ko', markersize = 8)

# labels etc
ax12.set_xlabel("S", fontsize = 14)
ax12.set_ylabel("I", fontsize = 14)
ax12.set_title("Phase plane for SIR model", fontsize = 16)

plt.show()

print(f'The number of infected individuals reaches a maximum of {int(np.floor(Imax))} rounded to the nearest integer.')

# set up grid for betas
nbetas = 100
Imax_list = []
betas = np.linspace(0, 5*(10**-4), nbetas)

# repeat analysis for range of betas
for i in range(0,nbetas):
    # call ODE solver
    y_sol = odeint(SIR, y0, t, (betas[i],))

    # extract I from the solution array y
    I = y_sol[:,1]
    Imax = I.max()
    Imax_list.append(Imax)

# setting up figure and subplot for second plot
fig2, ax21 = plt.subplots(figsize=(6,5))

# plot max of I against beta
ax21.plot(betas,np.array(Imax_list))
ax21.set_xlabel("$\\beta$", fontsize=14)
ax21.set_ylabel("Maximum of I", fontsize=14)
ax21.set_title("Maximum values of I against $\\beta$", fontsize=16)

plt.show()

```
